### 生成验证集的YOLO标签

```python
from pylabel import importer
```

```python
path_to_annotations = '/mnt/sdb/public/data/coco2017/annotations/instances_val2017.json'
path_to_images = '/mnt/sdb/public/data/coco2017/val2017'
```

```python
dataset = importer.ImportCoco(path_to_annotations, path_to_images=path_to_images, name="coco2017")
dataset.df.head(5)
```

### Analyze Annotations

```python
print(f"Number of images: {dataset.analyze.num_images}")
print(f"Number of classes: {dataset.analyze.num_classes}")
print(f"Classes:{dataset.analyze.classes}")
print(f"Class counts:\n{dataset.analyze.class_counts}")
print(f"Path to annotations:\n{dataset.path_to_annotations}")
```
